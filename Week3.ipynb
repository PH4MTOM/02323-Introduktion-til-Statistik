```python
## Plotting 1

# Define the range for x-axis
x_1 = np.linspace(-4, 4, 1000)
pdf_1 = stats.norm.pdf(x_1)

# Plot the normal distribution curve
plt.plot(x_1, pdf_1, label='Standard Normal Distribution (=0, =1)', color='blue')

# Shade the area under the curve up to x = 2
x_fill_1 = np.linspace(-4, 2, 1000)
plt.fill_between(x_fill_1, stats.norm.pdf(x_fill_1), color='skyblue', alpha=0.5, label='P(X  2)')

# Add a vertical line at x = 2
plt.axvline(x=2, color='red', linestyle='--', label='x = 2')

# Annotate the CDF value
cdf_value_1 = stats.norm.cdf(2)
plt.text(2, 0.1, f'CDF(2)  {cdf_value_1:.3f}', color='red', ha='left')

# Final plot settings
plt.title('1 - Standard Normal Distribution with Shaded CDF Area')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

## Plotting 2 (doing the same as plot 1)

# Define the range for x-axis
x_2 = np.linspace(-3, 5, 1000)
pdf_2 = stats.norm.pdf(x_2, 1, 1)

# Plot the normal distribution curve
plt.plot(x_2, pdf_2, label='Normal Distribution (=1, =1)', color='blue')

# Shade the area under the curve up to x = 2
x_fill_2 = np.linspace(-3, 2, 1000)
plt.fill_between(x_fill_2, stats.norm.pdf(x_fill_2, 1, 1), color='skyblue', alpha=0.5, label='P(X  2)')

# Add a vertical line at x = 2
plt.axvline(x=2, color='red', linestyle='--', label='x = 2')

# Annotate the CDF value
cdf_value_2 = stats.norm.cdf(2, 1, 1)
plt.text(2, 0.35, f'CDF(2, 1, 1)  {cdf_value_2:.3f}', color='red', ha='left')

# Final plot settings
plt.title('2 - Standard Normal Distribution with Shaded CDF Area')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

## Plotting 3

x_3 = np.linspace(-6, 8, 1000)
# The Python commando takes in the standard deviation (sigma).
pdf_3 = stats.norm.pdf(x_3, 1, 2)

plt.plot(x_3, pdf_3, label='Normal Distribution (=1, =2)', color='blue')

x_fill_3 = np.linspace(-6, 2, 1000)
plt.fill_between(x_fill_3, stats.norm.pdf(x_fill_3, 1, 2), color='skyblue', alpha=0.5, label='P(X  2)')

plt.axvline(x=2, color='red', linestyle='--', label='x = 2')

cdf_value_3 = stats.norm.cdf(2, 1, 2)
plt.text(2, 0.2, f'CDF(2, 1, 2)  {cdf_value_3:.3f}', color='red', ha='left')

plt.title('3 - Standard Normal Distribution with Shaded CDF Area')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

##### b) Calculate the following probabilities

Let X be normally distributed with mean 24 and variance 16



The mean $\mu = 24$ and the variance $\sigma^2 = 4$

P(X <= 20):

Instead of using the CDF formula for normal distribution, we can transform X into a standardized normal random variable \
with the following formula

$$Z = \frac{X - \mu}{\sigma}$$

We then calculate the Z-score and make a lookup in a t-table and get the CDF value.

See the following calculations:

Calculation the standard deviation
$$\sigma = \sqrt{\sigma^2} = \sqrt{4^2} = 4$$

Insert the known variables into the Z formula

$$Z = \frac{X - \mu}{\sigma} = \frac{20 - 24}{4} = -\frac{4}{4} = -1$$

Doing a Z-table lookup and we get the following https://math.arizona.edu/~rsims/ma464/standardnormaltable.pdf

$$ P(X < 20) = P(Z < -1) = 0.1586 $$

P(X > 29.5):

To find P(X > 29.5), we can write the formula as

$$P(X > 29.5) = 1 - P(X < 29.5)$$

We are doing the same method as the last calculation, using the T-table to find the value of CDF.

Insert the known values (we know the std from the last calculation to be 4)

$$Z = \frac{X - \mu}{\sigma} = \frac{29.5 - 24}{4} = \frac{5.5}{4} = 1.375$$

Since our Z-table only covers 1.37 and 1.38, we need to take the values from both and calculate the different between them and find the z-value for 1.375. See the calculations below

$$z_{between} = \frac{z_{1.37} - z_{1.38}}{2} = \frac{0.91621 - 0.91466}{2} = \frac{0.00155}{2} = 0.000775$$
$$z_{1.375} = z_{1.37} + z_{between} = 0.91466 + 0.000775 = 0.915435$$

$$P(X > 29.5) = 1 - P(Z < 1.375) = 1 - 0.915435 = 0.08456$$

P(X = 23.8):

The probability of a continuous random variable to have the outcome equal to a single value is zero. So $P(X = 23.8) = 0$


```python
# P(X < 20)
print(f'P(X < 20) = {stats.norm.cdf(20, loc=24, scale=4)}')

# P(X > 29.5)
print(f'P(X > 29.5) = {1 - stats.norm.cdf(29.5, loc=24, scale=4)}')
```

##### b) What is the mean () and standard deviation () for the total time use for checking 100 chips on the new machine is?

We have been given a normal distribution that represents time for 1 check. In this task, they want us to create a new normal distribution by finding the mean and standard deviation for time for 100 checks.

To find these, we will use theorem 2.56 from Linear Combination for mean and variance.



Calculations:

The new machine: $\mu = 54$, $\sigma^2 = 3^2$

$n = 100$

$\mu = E(U) = E(\sum^{100}_{i=1} X_i) = \sum^{100}_{i=1} 54 = 100 * 54 = 5400$

$\sigma^2 = V(U) = U(\sum^{100}_{i=1} X_i) = \sum^{100}_{i=1} 3^2 = \sum^{100}_{i=1} 9 = 100 * 9 = 900$

Finding the standard deviation (std)

$\sigma = \sqrt{\sigma^2} = \sqrt{900} = 30$

We have now scaled the normal distribution for 100 checks instead of 1, and we found the following mean and standard deviation

Mean: $\mu = 5400$, Std: $\sigma = 30$

The new normal distribution would be $N(5400, 30^2)$

##### a) The expected error rate in the manufacturing will be?

The task is describing a normal distribution with mean $\mu_L = 3000mm$ and standard deviation (std) $\sigma_L = 3 mm$.

These let us describe the following normal distribution $N(3000, 3^2)$

We then get the thresholds the items has to be inside for not to be considered an error. The item should not be more than 3007 mm and length most be 2993 mm.

Since we have a normal distribution, we can calculate the probabilities for these intervals that then gives us the probability of error. See below:

Expected Error Rate = $P(L \leq 2993) + P(L \geq 3007)$

Lets calculate the first term $P(L \leq 2993)$

We will use the handy transformation formel

$Z = \frac{X - \mu}{\sigma} = \frac{2993 - 3000}{3} = \frac{-7}{3} = -2.33$

Lookup in our t-table https://math.arizona.edu/~rsims/ma464/standardnormaltable.pdf and we get 

$P(L \leq 2993) = 0.0099$

Calculating $P(L \geq 3007)$

$P(L \geq 3007) = 1 - P(L \leq 3007)$

$Z = \frac{X - \mu}{\sigma} = \frac{3007 - 3000}{3} = \frac{7}{3} = 2.33$

$P(L \geq 3007) = 1 - P(L \leq 3007) = 1 - 0.9901 = 0.0099

We then add the number together to find our error rate

Expected Error Rate = $P(L \leq 2993) + P(L \geq 3007) = 0.0099 + 0.0099 = 0.0198$

The result in procent is

$100 * 0.0198 = 1.98%$

The error rate is around 2%
