##### a) The following Python commands and results are given:

1 - `print((stats.norm.cdf(2))` = 0.9772498680518208

2 - `print(stats.norm.cdf(2,1,1))` = 0.8413447460685429

3 - `print(stats.norm.cdf(2,1,2))` = 0.6914624612740131

*Specify which distributions are used and explain the resulting probabilities (preferably by a sketch).*

If we look up the Python commando, we see that the commando represent the Normal Distribution using CDF.

How to read the Python Commando:

cdf(x, loc=0, scale=1), loc is the mean $\mu$, and scale is the standard deviancen $\sigma$. NOTE: Because we are using a CDF, its P(X <= x).

Looking at 1, because the mean and variance hasn't been added to the Commando, it will follow the standard normal distribution $N(0,1)$. We read the following result:

$$P(X <= 2) = 0.9772 \;\; for \;\; N \sim (\mu = 0, \sigma^2 = 1)$$

Looking at 2, We read the following result:

$$P(X <= 2) = 0.8413 \;\; for \;\; N \sim (\mu = 1, \sigma^2 = 1)$$

Looking at 3, we read the following result:

$$P(X <= 2) = 0.6914 \;\; for \;\; N \sim (\mu = 1, \sigma^2 = 4)$$

NOTE: The commando is using standard deviation $\sigma = 2$, but we are showing it in $\sigma^2$, sÃ¶ $2^2$ = 4

Sketch will be added later

```python
## Plotting 1

# Define the range for x-axis
x_1 = np.linspace(-4, 4, 1000)
pdf_1 = stats.norm.pdf(x_1)

# Plot the normal distribution curve
plt.plot(x_1, pdf_1, label='Standard Normal Distribution (Î¼=0, Ï=1)', color='blue')

# Shade the area under the curve up to x = 2
x_fill_1 = np.linspace(-4, 2, 1000)
plt.fill_between(x_fill_1, stats.norm.pdf(x_fill_1), color='skyblue', alpha=0.5, label='P(X â¤ 2)')

# Add a vertical line at x = 2
plt.axvline(x=2, color='red', linestyle='--', label='x = 2')

# Annotate the CDF value
cdf_value_1 = stats.norm.cdf(2)
plt.text(2, 0.1, f'CDF(2) â {cdf_value_1:.3f}', color='red', ha='left')

# Final plot settings
plt.title('1 - Standard Normal Distribution with Shaded CDF Area')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

## Plotting 2 (doing the same as plot 1)

# Define the range for x-axis
x_2 = np.linspace(-3, 5, 1000)
pdf_2 = stats.norm.pdf(x_2, 1, 1)

# Plot the normal distribution curve
plt.plot(x_2, pdf_2, label='Normal Distribution (Î¼=1, Ï=1)', color='blue')

# Shade the area under the curve up to x = 2
x_fill_2 = np.linspace(-3, 2, 1000)
plt.fill_between(x_fill_2, stats.norm.pdf(x_fill_2, 1, 1), color='skyblue', alpha=0.5, label='P(X â¤ 2)')

# Add a vertical line at x = 2
plt.axvline(x=2, color='red', linestyle='--', label='x = 2')

# Annotate the CDF value
cdf_value_2 = stats.norm.cdf(2, 1, 1)
plt.text(2, 0.35, f'CDF(2, 1, 1) â {cdf_value_2:.3f}', color='red', ha='left')

# Final plot settings
plt.title('2 - Standard Normal Distribution with Shaded CDF Area')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

## Plotting 3

x_3 = np.linspace(-6, 8, 1000)
# The Python commando takes in the standard deviation (sigma).
pdf_3 = stats.norm.pdf(x_3, 1, 2)

plt.plot(x_3, pdf_3, label='Normal Distribution (Î¼=1, Ï=2)', color='blue')

x_fill_3 = np.linspace(-6, 2, 1000)
plt.fill_between(x_fill_3, stats.norm.pdf(x_fill_3, 1, 2), color='skyblue', alpha=0.5, label='P(X â¤ 2)')

plt.axvline(x=2, color='red', linestyle='--', label='x = 2')

cdf_value_3 = stats.norm.cdf(2, 1, 2)
plt.text(2, 0.2, f'CDF(2, 1, 2) â {cdf_value_3:.3f}', color='red', ha='left')

plt.title('3 - Standard Normal Distribution with Shaded CDF Area')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
