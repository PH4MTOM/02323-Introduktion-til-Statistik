If we look up the Python commando, we see that the commando represent the Normal Distribution using CDF.

###### How to read the Python Commando:

cdf(x, loc=0, scale=1), loc is the mean $\mu$, and scale is the standard deviancen $\sigma$. NOTE: Because we are using a CDF, its P(X <= x).

Looking at 1, because the mean and variance hasn't been added to the Commando, it will follow the standard normal distribution $N(0,1)$. We read the following result:

$$P(X <= 2) = 0.9772 \;\; for \;\; N \sim (\mu = 0, \sigma^2 = 1)$$

Looking at 2, We read the following result:

$$P(X <= 2) = 0.8413 \;\; for \;\; N \sim (\mu = 1, \sigma^2 = 1)$$

Looking at 3, we read the following result:

$$P(X <= 2) = 0.6914 \;\; for \;\; N \sim (\mu = 1, \sigma^2 = 4)$$

NOTE: The commando is using standard deviation $\sigma = 2$, but we are showing it in $\sigma^2$, sÃ¶ $2^2$ = 4
