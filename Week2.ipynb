#### 2.4 - Consumer survey

In a consumer survey performed by a newspaper, 20 different groceries (products) were purchased in a grocery store. Discrepancies between the price appearing on the sales slip and the shelf price were found in 6 of these purchased
products.

##### a) At the same time a customer buys 3 random (different) products within the group consisting of the 20 goods in the store. The probability that no discrepancies occurs for this customer is?

First we analyze what distribution we would use for this scenario. The Poission Distribution would not make sense here, as that used to represent probability over intervals and here we have none. It would also not make sense to use the Binomial distribution as the customer picks 3 **different** products, which means without replacement. We therefore conclude to use the Hypergeometric Distribution.

- n = Number of draws,
- a = Number of successes in population,
- N = Population Size,
- x = Number of successes in sample.

$$f(x;n,a,N) = P(X = x) = \frac{\binom{a}{x} \binom{N - a}{n - x}}{\binom{N}{n}}$$

$$\binom{a}{b} = \frac{a!}{b!(a-b)!'}$$

The customer buy 3 products (n = 3). We have a total of 20 different products (N = 20). We have 6 discrepancies prices found (a = 6). We want to calculate the probability for the customer to have bought 0 discrepanies products (x = 0). We will insert into the formula with our observations.

$$f(x;n,a,N) = P(X = 0) = \frac{\binom{6}{0} \binom{20 - 6}{3 - 0}}{\binom{20}{3}} = \frac{\binom{6}{0} \binom{14}{3}}{\binom{20}{3}} \leftrightarrow$$

Calculating the different binoms with the formula above:

$$\binom{6}{0} = \frac{6!}{0! (6-0)!} = \frac{6!}{6!} = 1$$

$$\binom{14}{3} = \frac{14!}{3! (14 - 3)!} = \frac{14!}{3! \cdot 11!} = \frac{87.178.291.200}{6 \cdot 39.916.800} = \frac{87.178.291.200}{239.500.800} = 364$$

$$\binom{20}{3} = \frac{20!}{3! (20 - 3)!} = \frac{20!}{3! \cdot 17!} = \frac{2.432.902.008.000.000.000}{6 \cdot 355.687.428.100.000} = \frac{2.432.902.008.000.000.000}{2.134.124.569.000.000} = 1140$$

Insert the binoms inside the formula and we get

$$f(x;n,a,N) = P(X = 0) = \frac{\binom{6}{0} \binom{14}{3}}{\binom{20}{3}} = \frac{1 \cdot 364}{1140} = \frac{364}{1140} = \underline{\underline{\frac{91}{285} \approx 0.3192}}$$

We conclude that the probability for the customer to have bought 0 discrepanies products are 0.3192 * 100 = 31.92%


#### 2.5 - Hay delivery quality

A horse owner receives 20 bales of hay in a sealed plastic packaging. To control the hay, 3 bales of hay are randomly selected, and each checked whether it contains harmful fungal spores.

It is believed that among the 20 bales of hay 2 bales are infected with fungal spores. A random variable X describes the number of infected bales of hay among the three selected.

##### a) The mean of X ($\mu_{x}$), the variance of X $(\sigma^{2}_{x})$ and P(X >= 1) are?

To figure out the formulas for the mean and variance, we need to determine what distribution is the correct one to use for this math problem. Again the Poisson distribution is not the correct one to use, because our problem has nothing to do with intervals over time. We then look at the Binomial distribution or Hypergeometric distribution. Since we have without replacement in this problem, we pick 3 randomly selected haybales and therefore will not check them again, we have to use the Hypergeometric distribution. 

See 2.4 for how the hypergeometric distribution looks.

We then analyse from the math problem that we have a population of 20 haybales (N = 20). We pick 3 random selected haybales (n = 3). The number of infected hay bales are 2 (a = 2).

Summary: N = 20, n = 3, a = 2 and x = 0 (will explain later).

To calculate the mean and variance are we using the following formula for Hypergemo:

Mean:

$$\mu = n \frac{a}{N} = 3 \cdot \frac{2}{20} = \frac{3 \cdot 2}{20} = \frac{6}{20} = \underline{\underline{\frac{3}{10} \approx 0.3}}$$

Variance:

$$
\begin{aligned}
\sigma^2 = n \frac{a(N-a)}{N^2} \frac{N - n}{N - 1} = 3 \cdot \frac{2 (20-2)}{20^2} \cdot \frac{20-3}{20-1} &= \\ \\
3 \cdot \frac{2 \cdot 18}{400} \cdot \frac{20-3}{20-1} = 3 \cdot \frac{36}{400} \cdot \frac{20-3}{20-1} &= \\ \\
\frac{3 \cdot 36}{400} \cdot \frac{20-3}{20-1} = \frac{108}{400} \cdot \frac{20-3}{20-1} &= \\ \\
\frac{108}{400} \cdot \frac{17}{19} = 0.27 \cdot 0.8947 &= \underline{\underline{0.2415}}
\end{aligned}
$$

To calculate the probability that there is one or more infected haybales, we first have to calculate the probability that no haybales have been infeted P(X = 0). 

Thats why we set x = 0 earlier. With this probability of no infested haybales, we cal calculate P(X >= 1) = 1 - P(x = 0).

We start of by inserting the values we gathered from the math problem:

$$P(X=0) = \frac{\binom{2}{0} \binom{20-2}{3-0}}{\binom{20}{3}} = \frac{\binom{2}{0} \binom{18}{3}}{\binom{20}{3}}$$

We then calculate each binom. Note: I am not including every intermediate calculation for this. See exercise 2.4 for deeper dive.

$$\binom{2}{0} = \frac{2!}{0!(2-0)!} = \frac{2!}{2!} = 1$$

$$\binom{18}{3} = \frac{18!}{3!(18-3)!} = \frac{18!}{3! \cdot 15!} = 816$$

$$\binom{20}{3} = \frac{20!}{3!(20-3)!} = \frac{20!}{3! \cdot 17!} = 1140$$

Insert the calculated binoms:

$$P(X=0) = \frac{\binom{2}{0} \binom{18}{3}}{\binom{20}{3}} = \frac{1 \cdot 816}{1140} = \frac{816}{1140} = 0.7157$$

We can then calculate P(X >= 1) as follows

$$P(X >= 1) = 1 - P(X=0) = 1 - 0.7157 = \underline{\underline{0.284}}$$

We conclude that there is a $0.284 \cdot 100 = 28.4%$ chance that there is 1 or more haybales infected with harmful fungal spores

```python
# How to calculate it in Python.

# Init the values from the exercise
N = 20
n = 3
a = 2
x = 0

# Calculate mean stats.hypergeom.mean(population, successes, draws)
mean = stats.hypergeom.mean(N, a, n)
print(f'Hypergeom Mean: {mean}')

# Calculate variance stats.hypergeom.var(population, successes, draws)
variance = stats.hypergeom.var(N, a, n)
print(f'Hypergeom Variance: {variance}')

# Calculate Probability P(X >= 1)

# P(X = 0)
prob_a = stats.hypergeom.pmf(x, N, a, n)

# P(x >= 1)
prob_b = (1 - prob_a)

print(f'P(X >= 1): {prob_b * 100}')
```

#### 2.7 - A fully automated production

On a large fully automated production plant items are pushed to a side band at random time points, from which they are automatically fed to a control unit. The production plant is set up in such a way that the number of items sent to
the control unit on average is 1.6 item pr. minute. Let the random variable X denote the number of items pushed to the side band in 1 minute. It is assumed that X follows a Poisson distribution.

##### a) What is the probability that there will arrive more than 5 items at the control unit in a given minute is?

Since we are working with a time interval we will use the Poisson Distribution that have the following formula for P(X = x)

$$f(x; \lambda) = P(X = x) = \frac{\lambda^x}{x!} e^{-\lambda}$$

where lambda is the time interval (rate / intensity)

Reading from the text, we get the following variables:

$$\lambda = 1.6$$

To calculate the probability that there will arrive more than 5 items at the control unit in a given minute is

P(X > 5) = 1 - P(X >= 5)

P(X <= 5) = P(X = 0) + P(X = 1) + P(X = 2) + P(X = 3) + P(X = 4) + P(X = 5)

I am only showing detail calculations for P(X = 1). The rest is done through the calculator

$$
\begin{aligned}
P(X = 0) = \frac{1.6^0}{0!} e^{-1.6} &= 0.2018965 \\ \\
P(X = 1) = \frac{1.6^1}{1!} e^{-1.6} &= \frac{1.6}{1} e^{-1.6} = 1.6 \cdot e^{-1.6} = 1.6 0.201896 = 0.32303 \\ \\
P(X = 2) = \frac{1.6^2}{2!} e^{-1.6} &= 0.258427 \\ \\
P(X = 3) = \frac{1.6^3}{3!} e^{-1.6} &= 0.137828 \\ \\
P(X = 4) = \frac{1.6^4}{4!} e^{-1.6} &= 0.0551312 \\ \\
P(X = 5) = \frac{1.6^5}{5!} e^{-1.6} &= 0.017641
\end{aligned}
$$

Calculating the P(X >= 5)

$$P(X >= 5) = 0.2018965 + 0.32303 + 0.258427 + 0.137828 + 0.0551312 + 0.017641 = 0.994$$

Calculating P(X > 5)

$$P(X > 5) = 1 - P(X <= 5) = 1 - 0.994 = 0.006$$

There is a approximately 0.6% (0.006 * 100 = 0.6) probability that there arrive 5 or more items at the control unit in a given minute

```python
# How to calculate it in Python
print(1-stats.poisson.cdf(k=5,mu=1.6))
```
